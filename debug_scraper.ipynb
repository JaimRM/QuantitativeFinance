```python
def debug_scrape():
    driver = webdriver.Safari()
    driver.set_window_size(1280, 900)
    driver.get("https://www.tradingview.com/symbols/LSE-BARC/forecast/")
    time.sleep(8)

    # Save the full page source so we can inspect it
    with open("/Users/macbook/Downloads/page_debug.html", "w") as f:
        f.write(driver.page_source)

    # Print everything that looks like a number
    all_elements = driver.find_elements(By.XPATH, "//*[not(self::script)][not(self::style)]")
    for el in all_elements:
        try:
            text = el.text.strip()
            if re.match(r'^\d{2,5}(\.\d+)?$', text):
                print(f"Found: '{text}' | class: {el.get_attribute('class')}")
        except:
            pass

    driver.quit()

def main():
    debug_scrape()
```
