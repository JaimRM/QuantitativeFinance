```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# 1. Preparación de los datos
# Buscamos que aprenda que: [x, x+10, x+20] -> x+30
X_train = np.array([[10, 20, 30], [20, 30, 40], [30, 40, 50], [40, 50, 60]])
y_train = np.array([40, 50, 60, 70])

# Las LSTM necesitan una entrada 3D: [muestras, pasos de tiempo, características]
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))

# 2. Definición del modelo
model = Sequential([
    # Capa LSTM con 50 neuronas. input_shape = (pasos de tiempo, características)
    LSTM(50, activation='relu', input_shape=(3, 1)),
    # Capa de salida que devuelve 1 solo valor
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')

# 3. Entrenamiento
print("Entrenando el modelo...")
model.fit(X_train, y_train, epochs=1000, verbose=0)

# 4. Predicción de prueba
# Probamos con una secuencia que el modelo NO ha visto: [50, 60, 70] -> debería dar 80
x_input = np.array([50, 60, 70]).reshape((1, 3, 1))
prediccion = model.predict(x_input, verbose=0)

print(f"Entrada: [50, 60, 70]")
print(f"Predicción del modelo: {prediccion[0][0]:.2f}")
```
